## 🎯 **PHASE 5: Advanced Feature (Tip Percentage)**

### **Creating Tip Percentage Feature**

A new column, `tip_percentage`, is created to analyze tipping behavior relative to the total bill, calculated as `(tip / total_bill) × 100`.

### **Analysis of Tip Percentage**

Female customers, on average, tend to tip a slightly higher percentage (16.65%) compared to Male customers (15.77%). While the absolute tip amount might vary, the proportion of the bill they tip is marginally higher for females in this dataset.

Friday and Sunday have the highest average tipping percentages (approximately 16.99% and 16.69% respectively) compared to other days.

```python
df['tip_percentage'] = (df['tip'] / df['total_bill'])* 100
plt.figure(figsize=(10,6))

sns.set_style('darkgrid')
ax = sns.barplot(x='sex', y='tip_percentage', hue='sex', data=df, palette='viridis', errorbar='sd', legend=False) # Removed legend

# Add numerical labels on top of the bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f%%')

plt.title('Average Tip Percentage by Gender', fontsize=18)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Tip Percentage (%)', fontsize=14)
plt.ylim(0, df['tip_percentage'].max() * 1.1) # Adjust y-axis limit for labels

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))

sns.set_style('darkgrid')

ax =  sns.barplot(x='day', y='tip_percentage', data=df)

for container in ax.containers:
  ax.bar_label(container, fmt='%.2f%%')

plt.title("Day vs Tip_Percentage", fontsize=18)
plt.ylabel("tip_percentage", fontsize=14)
plt.xlabel('Day', fontsize=14)
plt.ylim(0, df['tip_percentage'].max() * 1.1)

plt.grid(axis='x', linestyle=':', alpha=0.7)
plt.tight_layout()
plt.show()
```
